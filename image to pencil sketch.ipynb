# **Image_to_pencil_Sketch_with_python**

```python
import cv2
from google.colab.patches import cv2_imshow 
```

```python
img = cv2.imread("labrador-retriever-1210559__480.jpg")
cv2_imshow(img)
cv2.waitKey(0)
```

```python
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
cv2.waitKey(0)
```

```python
invertedImage = 255 - grayImage
cv2_imshow(invertedImage)
cv2.waitKey()
```

```python
blurred = cv2.GaussianBlur(invertedImage, (21, 21), 0)
```

PENCIL SKETCH

```python
invertedBlurred = 255 - blurred
pencil_sketch = cv2.divide(grayImage, invertedBlurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
